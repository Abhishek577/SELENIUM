```python
#we will wait till all radio buttons are visible
wait=WebDriverWait(driver,21)
wait.until(ec.presence_of_all_elements_located((By.XPATH,"//span[@class='fl b ng-binding']")))
time.sleep(5)
```

```python
#creating a dictionary to save the data
hotel={}

hotel['hotel_name']=[]

hotel['area']=[]

hotel['price']=[]

hotel['rating']=[]
```

```python
hotel
```

```python
#we will scroll down to the bottom so that page can be loaded properly\
time.sleep(3)
driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
time.sleep(12)
```

```python
#scrapping the names
name=driver.find_elements_by_xpath("//h2[@class='hotel-name full fs-18 three-dot']//a")
```

```python
print("no of hotels found =", len(name))
```

```python
for n in name:
    names=n.get_attribute("title")
    hotel['hotel_name'].append(names)
```

```python
hotel['hotel_name']
```

```python
#scrapping the location
loc=driver.find_elements_by_xpath("//span[@class='va-sub ng-binding']")
hotel['area']=[i.text for i in loc]
print("area:",len(hotel['area']) )                            
```

```python
hotel['area']
```

```python
#scrapping the price
price=driver.find_elements_by_xpath("//span[@class='bold ng-binding']")  #//p[@class='hotel-price-value hand']
h=[p.text for p in price]
hotel['price']=[i for i in h if i != '']
```

```python
len(hotel['price'])
```

```python
hotel['price']
```

```python
rating=driver.find_elements_by_xpath("//span[@class='mr-5']//span")
rate=[r.text for r in rating]
print(len(rate))
rate
```

```python
hotel['rating']=[i for i in rate if i != '']
```
